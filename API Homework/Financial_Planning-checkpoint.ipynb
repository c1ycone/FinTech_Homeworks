### Collect Crypto Prices Using the requests Library

```python
# Set current amount of crypto assets
my_btc = 1.2
my_eth = 5.3
```

```python
# Crypto API URLs
btc_url = "https://api.alternative.me/v2/ticker/Bitcoin/?convert=CAD"
eth_url = "https://api.alternative.me/v2/ticker/Ethereum/?convert=CAD"

# Execute GET requests
btc_response_data = requests.get(btc_url)
eth_response_data = requests.get(eth_url)

print(btc_response_data)
print(eth_response_data)
```

```python
 # Store response using `content` attribute
btc_response_content = btc_response_data.content
eth_response_content = eth_response_data.content

# Format data as JSON
btc_data = btc_response_data.json()
eth_data = eth_response_data.json()

# Use json.dumps with argument indent=4 to format data
print(json.dumps(btc_data, indent=5))
print(json.dumps(eth_data, indent=5))
```

```python
# Fetch current BTC price
btc_price = btc_data["data"]["1"]["quotes"]["CAD"]["price"]

# Fetch current ETH price
eth_price = eth_data["data"]["1027"]["quotes"]["CAD"]["price"]

# Print AGG and SPY close prices
print(f"Current BTC price: ${btc_price}")
print(f"Current ETH price: ${eth_price}")
```

```python
# Compute current value of my crpto
my_btc_value = my_btc * btc_price
my_eth_value = my_eth * eth_price
```

```python
# Print current crypto wallet balance
print(f"The current value of your {my_btc} BTC is ${my_btc_value:0.2f}")
print(f"The current value of your {my_eth} ETH is ${my_eth_value:0.2f}") 
```

### Collect Investments Data Using Alpaca: SPY (stocks) and AGG (bonds)

```python
# Current amount of shares
my_spy = 50
my_agg = 200

# Set current amount of shares data
shares_data = {
    "shares": [my_spy, my_agg]
}

# Set the tickers
tickers = ["SPY", "AGG"]

# Create the shares DataFrame with `tickers` as index
df_shares = pd.DataFrame(shares_data, index=tickers)

# Preview DataFrame
df_shares
```

```python
 # Create the Alpaca API object
alpaca = tradeapi.REST(
    alpaca_api_key,
    alpaca_secret_key,
    api_version="v2")
```

```python
# Format current date as ISO format
today = pd.Timestamp("2021-03-12", tz="America/New_York").isoformat()

 # Set timeframe to one day ('1D') for the Alpaca API
timeframe = "1D"

# Get current closing prices for SPY and AGG
df_portfolio = alpaca.get_barset(
    tickers,
    timeframe,
    start = today,
    end = today
).df

# Preview DataFrame
df_portfolio
```

```python
# Fetch the current closing prices from the DataFrame
agg_price = float(df_portfolio["AGG"]["close"])
spy_price = float(df_portfolio["SPY"]["close"])


# Print AGG and SPY close prices
print(f"Current AGG closing price: ${agg_price}")
print(f"Current SPY closing price: ${spy_price}")
```

```python
# Compute the current value in dollars of the stock portfolio
agg_value = agg_price * df_shares.loc["AGG"]["shares"]
spy_value = spy_price * df_shares.loc["SPY"]["shares"]

# Compute the value in dollars of the current amount of shares and print the results.
df_shares['values']=[agg_value, spy_value]
df_shares
```

```python
# Print the current value of the stocks portfolio
print(f"The current value of the {df_shares.loc['SPY']['shares']} SPY shares is ${spy_value:0.2f}.")
print(f"The current value of the {df_shares.loc['AGG']['shares']} AGG shares is ${agg_value:0.2f}.")
```

### Savings Health Analysis

```python
# Set monthly household income
monthly_income = 12000

# Create savings DataFrame
crypto_value = my_btc_value + my_eth_value
shares_value = spy_value + agg_value

savings_data = {
    "amount": [crypto_value, shares_value]
}

# Set the tickers
savings = ["crypto", "shares"]

# Create the shares DataFrame with `tickers` as index
df_savings = pd.DataFrame(savings_data, index=savings)


# Display savings DataFrame
display(df_savings)
```

```python
# Plot savings pie chart
df_savings.plot.pie(y="amount", title="Composition of Personal Savings")
```

```python
# Set ideal emergency fund
emergency_fund = monthly_income * 3

# Calculate total amount of savings
total_savings = crypto_value + shares_value

# Calculate savings needed
savings_needed = emergency_fund - total_savings

# Validate saving health
if emergency_fund  < total_savings:
    print("Congratulations! You have enough money in your emergency fund.")
elif total_savings == emergency_fund:
    print("Congratulations! You have reached your financial goal.")
else:
    print(f"You can do this! You need to save ${savings_needed:0.2f} to reach your financial goal.")
```

### Monte Carlo Simulation

```python
# Create the Alpaca API object
api = tradeapi.REST(
    alpaca_api_key,
    alpaca_secret_key,
    api_version="v2")

# Set start and end dates of five years back from today.
start_date = pd.Timestamp("2016-03-12", tz="America/New_York").isoformat()
end_date = pd.Timestamp("2021-03-12", tz="America/New_York").isoformat()

# Get 5 years' worth of historical data for SPY and AGG
df_ticker = api.get_barset(
    tickers,
    timeframe,
    start=start_date,
    end=end_date,limit=1000
).df

# Display sample data
df_ticker.head()
```

```python
# Print the documentation of the MCSimulation module 
?MCSimulation
```

```python
# Configure MC simulation to forecast 30 years cumulative returns
MC_30year = MCSimulation(
    portfolio_data = df_ticker,
    weights = [0.6, 0.4],
    num_simulation = 500,
    num_trading_days = 252*30
)

# Preview data
MC_30year.portfolio_data.head()
```

```python
# Run an MC simulation to forecast 30 years cumulative returns
MC_30year.calc_cumulative_return()
```

```python
# Plot simulation outcomes
line_plot = MC_30year.plot_simulation()

# Save the plot for future usage
line_plot.get_figure().savefig("MC_30year_sim_plot.png", bbox_inches="tight")
```

```python
#Plot probability distribution and confidence intervals
dist_plot = MC_30year.plot_distribution()

# Save the plot for future usage
dist_plot.get_figure().savefig('MC_30year_dist_plot.png',bbox_inches='tight')
```

### Early Retirement

```python
# Fetch summary statistics from the Monte Carlo simulation results
tbl = MC_30year.summarize_cumulative_return()

# Print summary statistics
print(tbl)
```

### Retirement Analysis

```python
# Use the lower and upper `95%` confidence intervals to calculate the range of the possible outcomes of initial $20,000 investment
ci_lower = round(tbl[8]*20000,2)
ci_upper = round(tbl[9]*20000,2)

# Print results
print(f"There is a 95% chance that an initial investment of $20,000 in the portfolio"
      f" over the next 30 years will end within in the range of"
      f" ${ci_lower} and ${ci_upper}")
```

```python
# Use the lower and upper `95%` confidence intervals to calculate the range of the possible outcomes on a 50% increase in the initial investment
ci_lower = round(tbl[8]*10000,2)
ci_upper = round(tbl[9]*10000,2)

# Print results
print(f"There is a 95% chance that an additional investment of $10,000 in the portfolio"
      f" over the next 30 years will end within in the range of"
      f" ${ci_lower} and ${ci_upper}")
```
